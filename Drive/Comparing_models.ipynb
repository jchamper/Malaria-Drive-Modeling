```python
# load deep learnign models
class AveragedModel:
    def __init__(self, model_files):
        self.models = [load_model(file) for file in model_files]

    def predict(self, input_data):
        # gain predictions for all models
        predictions = np.array([model.predict(input_data) for model in self.models])
        # calculate average number
        return np.mean(predictions, axis=0)
```

```python
# this function use 50% to split the success/ failed area and giving basic predictions
def check_accuracy(x,y):
    y=y.clip(0,1)
    """
        Check errors between predicted values and known result values.
        x: target value
        y: predicted value
    """
    rmse = np.sqrt(sum((x - y)**2 / len(x)))
    mae =sum(np.abs(x - y) / len(x))
    actual_success = 0
    true_positives = 0
    false_negatives = 0
    false_positives = 0
    for i in range(len(x)):
        if x[i] >= 0.5:
            actual_success += 1
        if y[i] >= 0.5 and x[i] >= 0.5:
            true_positives += 1
        if y[i] < 0.5 and x[i] >= 0.5:
            false_negatives += 1
        if y[i] >= 0.5 and x[i] < 0.5:
            false_positives += 1
    precision = 0
    recall = 0
    f1 = 0
    if true_positives + false_positives:
        precision = true_positives / (true_positives + false_positives)
    if actual_success:
        recall = true_positives / actual_success
    if precision + recall:
        f1 = 2 * precision *recall / (precision + recall)
    return rmse, mae,precision, recall, f1
```

```python
import math
def check_stat_accuracy(x,y,threshod=0.05):
    x=list(x)
    y=list(y)
    # the data collected from slim script will have random error
    # so even the best machine learning model could not be able to get precise estimation
    # then evaluate how many observation is located in the confidence interval created by y
    """
        Check errors between predicted values and known result values.
        x: target value
        y: predicted value
    """
    good_estimation=0
    bad_estimation=0
    control=0
    for i in range(len(x)):
            p=scipy.stats.binomtest(round(x[i]*10),10,y[i],'two-sided').pvalue
            
            # use random sampling as a control because the probability is discrete
            # the expected probability(y[i]) is regarded as the true probability
            # the number is regarded as a 10-repeated random experiment which follows binominal distribution
            
            if p<=threshod:
                bad_estimation+=1
            else:
                good_estimation+=1
            for j in range(1):
                p_CK=scipy.stats.binomtest(sum(np.random.random(10)<y[i]),10,y[i],'two-sided').pvalue
                if p_CK<=threshod:
                    control+=1
                    
    n=bad_estimation+good_estimation
    # accuracy(in CI) , all error- simulated_random_error= nonrandom error , n 
    return good_estimation/n,(bad_estimation-control)/n,n
```

```python
# the accuracy of models with different layers
def success_rate(obs,pred):
    actual_suppress=np.average(obs[obs>0])
    actual_all=np.average(obs)
    pred_suppress=np.average(pred[obs>0])
    pred_all=np.average(pred)
    return actual_all,pred_all,actual_suppress,pred_suppress
def get_accuracys(predict_model,predict_model2):
        elim_accuracy = check_accuracy(y_test["malaria_eliminated"], predict_model[0].flatten().clip(0,1))
        supp_accuracy = check_accuracy(y_test["suppressed"], predict_model2[0].flatten().clip(0,1))

        prev_r2 = r2_score(y_test["malaria_prevalence"], predict_model[1].flatten().clip(0,1))
        dens_r2 = r2_score(y_test["mosquito_prev"], predict_model2[1].flatten().clip(0,1))
        prev_rmse=math.sqrt(np.mean((y_test["malaria_prevalence"]-predict_model[1].flatten().clip(0,1))**2))
        dens_rmse=math.sqrt(np.mean((y_test["mosquito_prev"]-predict_model2[1].flatten().clip(0,1))**2))
        prev_mae=np.mean(np.abs(y_test["malaria_prevalence"]-predict_model[1].flatten().clip(0,1)))
        dens_mae=np.mean(np.abs(y_test["mosquito_prev"]-predict_model2[1].flatten().clip(0,1)))
        
        elim_ci1 = check_stat_accuracy((y_test[y_test["malaria_eliminated"]>0])["malaria_eliminated"], predict_model[0].flatten().clip(0,1)[y_test["malaria_eliminated"]>0],threshod=0.05)
        supp_ci1 = check_stat_accuracy((y_test[y_test["suppressed"]>0])["suppressed"], predict_model2[0].flatten().clip(0,1)[y_test["suppressed"]>0],threshod=0.05)
        elim_ci0 = check_stat_accuracy((y_test[y_test["malaria_eliminated"]==0])["malaria_eliminated"], predict_model[0].flatten().clip(0,1)[y_test["malaria_eliminated"]==0],threshod=0.05)
        supp_ci0 = check_stat_accuracy((y_test[y_test["suppressed"]==0])["suppressed"], predict_model2[0].flatten().clip(0,1)[y_test["suppressed"]==0],threshod=0.05)
        ma_obs_all,ma_pred_all,ma_obs_sup,ma_pred_sup=success_rate(y_test["malaria_eliminated"],predict_model[0].flatten().clip(0,1))
        mo_obs_all,mo_pred_all,mo_obs_sup,mo_pred_sup=success_rate(y_test["suppressed"],predict_model2[0].flatten().clip(0,1))
        return ((elim_accuracy[2],elim_accuracy[3],elim_accuracy[4],elim_ci1[0],elim_ci1[1],elim_ci0[0],elim_ci0[1],prev_rmse,prev_mae,prev_r2,ma_obs_all,ma_pred_all,ma_obs_sup,ma_pred_sup),
            (supp_accuracy[2],supp_accuracy[3],supp_accuracy[4],supp_ci1[0],supp_ci1[1],supp_ci0[0],supp_ci0[1],dens_rmse,dens_mae,dens_r2,mo_obs_all,mo_pred_all,mo_obs_sup,mo_pred_sup))
    
def total_accuracy():
    malaria=pd.DataFrame()
    mosquito=pd.DataFrame()
    malaria['lightBGM'],mosquito['lightBGM']=get_accuracys(lgb1,lgb2)
    malaria['XGBoost'],mosquito['XGBoost']=get_accuracys(xgb1,xgb2)
    malaria['CatBoost'],mosquito['CatBoost']=get_accuracys(cab1,cab2)
    #malaria['layer'+str(i)],mosquito['layer'+str(i)]=get_accuracys(predict_model,predict_model2)
    #malaria['layer'+str(i)],mosquito['layer'+str(i)]=get_accuracys(predict_model,predict_model2)
    for i,predict_model,predict_model2 in zip(range(1,6),predict_malaria,predict_mosquito): 
        print('malaria model accuracy: layer='+str(i))

        print('mosquito model accuracy: layer='+str(i))

        malaria['layer'+str(i)],mosquito['layer'+str(i)]=get_accuracys(predict_model,predict_model2)
    
    mosquito=mosquito.T
    malaria=malaria.T
    mosquito.columns=pd.MultiIndex.from_tuples([('mosquito distinction>=0.5','Precision'),('mosquito distinction>=0.5','Recall'),('mosquito distinction>=0.5','F1 value'),('distinction>0 in 95% CI','accuracy'),('distinction>0 in 95% CI','nonrandom error'),('distinction=0 in 95% CI','accuracy'),('distinction=0 in 95% CI','nonrandom error'),('mosquito density','RMSE'),('mosquito density','MAE'),('mosquito density','R square'),('total overall','oberved'),('total overall','predicted'),('suppress overall','observed'),('suppress overall','prediceted')])
    malaria.columns=pd.MultiIndex.from_tuples([('malaria elimination>=0.5','Precision'),('malaria elimination>=0.5','Recall'),('malaria elimination>=0.5','F1 value'),('elimination>0 in 95% CI','accuracy'),('elimination>0 in 95% CI','nonrandom error'),('elimination=0 in 95% CI','accuracy'),('elimination=0 in 95% CI','nonrandom error'),('malaria prevalence','RMSE'),('malaria prevalence','MAE'),('malaria prevalence','R square'),('total overall','oberved'),('total overall','predicted'),('suppress overall','observed'),('suppress overall','prediceted')])
    return mosquito,malaria
mosquito,malaria=total_accuracy()
mosquito.columns=pd.MultiIndex.from_tuples([('mosquito distinction>=0.5','Precision'),('mosquito distinction>=0.5','Recall'),('mosquito distinction>=0.5','F1 value'),('distinction>0 in 95% CI','accuracy'),('distinction>0 in 95% CI','nonrandom error'),('distinction=0 in 95% CI','accuracy'),('distinction=0 in 95% CI','nonrandom error'),('mosquito density','RMSE'),('mosquito density','MAE'),('mosquito density','R square'),('total overall','oberved'),('total overall','predicted'),('suppress overall','observed'),('suppress overall','prediceted')])
malaria.columns=pd.MultiIndex.from_tuples([('malaria elimination>=0.5','Precision'),('malaria elimination>=0.5','Recall'),('malaria elimination>=0.5','F1 value'),('elimination>0 in 95% CI','accuracy'),('elimination>0 in 95% CI','nonrandom error'),('elimination=0 in 95% CI','accuracy'),('elimination=0 in 95% CI','nonrandom error'),('malaria prevalence','RMSE'),('malaria prevalence','MAE'),('malaria prevalence','R square'),('total overall','oberved'),('total overall','predicted'),('suppress overall','observed'),('suppress overall','prediceted')])
print('mosquito model:')
display(mosquito)
print('malaria model:')
display(malaria)
```
